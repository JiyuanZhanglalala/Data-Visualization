## Choropleth

### Borough Level

The above data set does not have borough level info, so we extrat data from web

```python
table=pd.read_html('https://www.nycbynatives.com/nyc_info/new_york_city_zip_codes.php')
```

```python
Zip_Code_Borought=table[0]
```

```python
Zip_Code_Borought
```

```python
Zip_Code_Borought.columns=["Zip", "Borought", 2, "Zip", "Borought"]
```

```python
#Zip_Code_Borought[[3, 4]].rename(columns={3:0,4:1}) # Have to rename the last two columns in order to append them

Zip_Code_Borought=Zip_Code_Borought[[0, 1]].append(Zip_Code_Borought[[3, 4]].rename(columns={3:0,4:1}))
```

```python
Zip_Code_Borought.columns=["Zip","Borought"]
```

```python
df1=pd.merge(CVS_Stores,Zip_Code_Borought,left_on="PostalCode", right_on="Zip",how='inner')
```

```python
Count_of_Stores_in_Diff_Borought=df1.groupby(["Borought"]).count().reset_index()[["Borought","Line"]]
```

```python
Count_of_Stores_in_Diff_Borought.columns=['Borought', 'Count']
```

```python
Count_of_Stores_in_Diff_Borought
```

```python
NYC_Lat=40.7128
NYC_Long=-74.0060

NYC_Map=folium.Map(location=[NYC_Lat,NYC_Long],zoom_start=10)
```

```python
NYC_Map.choropleth(geo_data=GeoJsonFile,data=Count_of_Stores_in_Diff_Borought,columns=['Borought','Count'],key_on='feature.properties.borough',fill_color='YlOrBr',fill_opacity=0.7,line_opacity=0.2)
NYC_Map
```
